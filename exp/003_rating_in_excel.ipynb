# Exp3: Export to excel for manual rating

This file exports the augmented dataframe to an excel and reads it back in after manual evaluation.

```python
import pandas as pd
import sys
import os
sys.path.append(os.path.dirname(os.getcwd()))
import config

INPUT_PATH = '../dat/egpaugmented.json'
EXCEL_PATH = '../dat/augmented_examples_shuffled.xlsx'
JSON_PATH = '../dat/augmented_examples_shuffled.json'
```

Get the unrated. augmented dataset

```python
df = pd.read_json(INPUT_PATH)
```

```python
df.head()
```

```python
df.prompt.iloc[0]
```

Concat all examples, maintaining info if they are positive or negative examples.

```python
df['augmented_examples'] = df['augmented_examples'].apply(lambda examples: list(zip([True] * len(examples), examples)))
df['augmented_negative_examples'] = df['augmented_negative_examples'].apply(lambda examples: list(zip([False] * len(examples), examples)))
df['all_examples'] = df['augmented_examples'] + df['augmented_negative_examples']
df['all_examples']
```

Create one row per example

```python
new_df = df.explode('all_examples').drop(['augmented_examples', 'augmented_negative_examples', 'prompt'], axis=1, inplace=False)
new_df = new_df.groupby('#').apply(lambda x: x.sample(frac=1, random_state=config.SEED)).reset_index(drop=True)
new_df
```

Split back into two columns

```python
new_df[['positive', 'example']] = new_df['all_examples'].apply(pd.Series)
new_df.drop(['all_examples'], axis=1, inplace=True)
new_df.head()
```

Save the entire information into a json

```python
new_df.to_json(JSON_PATH)
```

Save into an excel file without the positive label

```python
new_df['rating'] = 0

new_df.drop('positive', axis=1).to_excel(EXCEL_PATH, index=False)
```

## Analyse results

After having rated with excel read in the columns

```python
rating_df = pd.read_excel(EXCEL_PATH)
rating_df.head()
```

Add information about the source of the example

```python
source_df = pd.read_json(JSON_PATH)
rating_df['source'] = source_df['positive']
rating_df.head()
```

Calculate the average correct rate per construction

```python
rating_df['source'] = rating_df['source'].astype(int) 
rating_df['rating'] = rating_df['rating'] - 1 
rating_df['correct'] = (rating_df['rating'] == rating_df['source'])
```

```python
rating_df.groupby('#')['correct'].mean()
```

```python
rating_df.groupby(['Level', 'Can-do statement', 'source'])['correct'].mean()
```
