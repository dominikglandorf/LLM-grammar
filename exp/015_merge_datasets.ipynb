```python
import pandas as pd
```

```python
cefr_levels = ['../dat/egp_A1.json', '../dat/egp_A2.json', '../dat/egp_B1.json', '../dat/egp_B2.json', '../dat/egp_C1.json', '../dat/egp_C2.json']
```

```python
main_df = pd.read_json('../dat/egp.json')
```

```python
for level in cefr_levels:
    df = pd.read_json(level)
    # filter df for non-empty arrays in column augmented_examples and add to main_df
    df = df[df['augmented_examples'].apply(lambda x: len(x) > 0)]
    # overwrite these rows in main_df
    main_df.update(df)
```

```python
# count array length in main_df augmented_examples and return the number of array larger than 250
print(main_df['augmented_examples'].apply(lambda x: len(set(x)) >= 500).sum())
print(main_df['augmented_negative_examples'].apply(lambda x: len(set(x)) >= 250).sum())
```
